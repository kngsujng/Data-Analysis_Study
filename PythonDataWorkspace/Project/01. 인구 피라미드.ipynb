# 1. 인구 피라미드

## 남자 데이터 정의

```python
import pandas as pd
df_m = pd.read_excel('201108_201108_연령별인구현황_월간.xlsx', skiprows=3, index_col='행정기관', usecols='B, E:Y')
df_m.head(3)
```

```python
df_m.iloc[0] = df_m.iloc[0].str.replace(',', '').astype(int) # 1,195,951 -> 1195951(정수형)
```

```python
df_m.iloc[0]
```

## 여자 데이터 정의

```python
df_f = pd.read_excel('201108_201108_연령별인구현황_월간.xlsx', skiprows=3, index_col='행정기관', usecols='B, AB:AV')
df_f
```

```python
df_m.columns
```

```python
df_f.columns
```

```python
df_f.columns = df_m.columns # 컬럼명 통일
df_f.columns
```

```python
df_f.iloc[0] = df_f.iloc[0].str.replace(',', '').astype(int)
```

```python
df_f.head(3)
```

## 데이터 시각화

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic' 
# matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상 해결
```

```python
plt.figure(figsize=(10,7))
plt.barh(df_m.columns, -df_m.iloc[0] // 1000 ) # 단위 : 천 명
plt.barh(df_f.columns, df_f.iloc[0] // 1000)
plt.title('2011 대한민국 인구 피라미드')
plt.savefig('2011_인구피라미드.png')
plt.show()
```
