# 2. 출생아 수 및 합계출산율

```python
import pandas as pd
df = pd.read_excel('출생아수및합계출산율.xlsx', skiprows=2, nrows=2, index_col=0)
# nrows가 2개인 이유 : column head 제외하고 data 2줄만 가져온다. 
df
```

```python
df.index
```

```python
df.index.values
```

```python
df.rename(index={'출생아\xa0수':'출생아 수', '합계\xa0출산율':'합계 출산율'}, inplace=True)
```

```python
df.loc['출생아 수']
```

```python
df.iloc[0] # 0번째 인덱스 값 가져오기 
```

```python
df = df.T # row와 column 바꾸기 
```

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic' 
# matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상 해결
```

```python
plt.plot(df.index, df['출생아 수'])
plt.plot(df.index, df['합계 출산율'])
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(10,5))
ax1.plot(df.index, df['출생아 수'], color='#ff812d')

ax2 = ax1.twinx() # x축을 공유하는 쌍둥이 axis
ax2.plot(df.index, df['합계 출산율'], color='#ffd100')
```

```python
fig, ax1 = plt.subplots(figsize=(13,5))
ax1.bar(df.index, df['출생아 수'], color='#ff812d')
ax1.set_ylim([250, 700])
ax1.set_yticks([300, 400, 500, 600])
ax1.set_ylabel('출생아 수 (천 명)')

ax2 = ax1.twinx() # x축을 공유하는 쌍둥이 axis
ax2.plot(df.index, df['합계 출산율'], color='#ffd100', marker='o')
ax2.set_ylim([0, 1.5])
ax2.set_yticks([0, 1])
ax2.set_ylabel('합계 출산율')

plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(13,5))
ax1.bar(df.index, df['출생아 수'], color='#ff812d')
ax1.set_ylim([250, 700])
ax1.set_yticks([300, 400, 500, 600])
ax1.set_ylabel('출생아 수 (천 명)')
for idx, val in enumerate(df['출생아 수']):
    ax1.text(idx, val+12, val, ha='center')

ax2 = ax1.twinx() # x축을 공유하는 쌍둥이 axis
ax2.plot(df.index, df['합계 출산율'], color='#ffd100', marker='o')
ax2.set_ylim([0, 1.5])
ax2.set_yticks([0, 1])
ax2.set_ylabel('합계 출산율')

plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(13,5))
ax1.bar(df.index, df['출생아 수'], color='#ff812d')
ax1.set_ylim([250, 700])
ax1.set_yticks([300, 400, 500, 600])
ax1.set_ylabel('출생아 수 (천 명)')
for idx, val in enumerate(df['출생아 수']):
    ax1.text(idx, val+12, val, ha='center')

ax2 = ax1.twinx() # x축을 공유하는 쌍둥이 axis
ax2.plot(df.index, df['합계 출산율'], color='#ffd100', marker='o', ms=15, lw=5, mec='w', mew=3)
ax2.set_ylim([0, 1.5])
ax2.set_yticks([0, 1])
ax2.set_ylabel('합계 출산율')

plt.show()
```
